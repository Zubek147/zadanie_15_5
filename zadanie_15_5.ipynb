```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
# Dane ceny produktu w PLN za każdy miesiąc
prices = [
    (1, 2.12),
    (2, 2.56),
    (3, 3.10),
    (4, 3.16),
    (5, 3.58),
    (6, 5.12),
    (7, 5.16),
    (8, 5.20),
    (9, 4.12),
    (10, 4.10),
    (11, 3.65),
    (12, 4.25),
]
```

```python
# Stworzenie obiektu DataFrame
df = pd.DataFrame(prices, columns=['Month', 'PricePLN'])
```

```python
# Ustawienie indeksu na miesiąc
df.set_index('Month', inplace=True)
```

```python
# Dodanie nowej kolumny priceUSD
df['PriceUSD'] = df['PricePLN'] / 4.0  # Przelicznik: 1 USD = 4 PLN
```

```python
# Wyświetlenie wyniku jako wykres liniowy koloru czerwonego z przerywaną linią
plt.plot(df.index, df['PriceUSD'], 'r--', dashes=(9,2), linewidth=1.9)
plt.xlabel('Month')
plt.ylabel('Price (USD)')
plt.title('Price of goods (USD)')
plt.yticks([i/4 for i in range(0, 9)])
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
